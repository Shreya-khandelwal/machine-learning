```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generate data using dummy function
X=np.arange(10)
y = (X-5)**2+3
print(X,y)
```

```python
# visualize 
plt.style.use('seaborn')
plt.plot(X,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# gradient descent
# init x with random value
plt.plot(X,y)
x = 0
lr = 0.1
for i in range(50):
    grad = 2*(x-5)
    x = x-lr*grad
    y = (X-5)**2+3
    plt.scatter(X,y)
    print(x)
plt.show()
```
